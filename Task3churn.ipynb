```python
import pandas as pd 
```

```python
train = pd.read_csv('dataset/Data Set For Task/Churn Prdiction Data/churn-bigml-20.csv')
test = pd.read_csv('dataset/Data Set For Task/Churn Prdiction Data/churn-bigml-80.csv')
train.head()
test.head()
```

```python
X_train = train.drop("Churn", axis=1)
y_train = train["Churn"]

X_test = test.drop("Churn", axis=1)
y_test = test["Churn"]
```

```python
train.dtypes
test.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['State'] = le.fit_transform(train['State'])
train['International plan'] = le.fit_transform(train['International plan'])
train['Voice mail plan'] = le.fit_transform(train['Voice mail plan'])

test['State'] = le.fit_transform(test['State'])
test['International plan'] = le.fit_transform(test['International plan'])
test['Voice mail plan'] = le.fit_transform(test['Voice mail plan'])
```

```python
train.select_dtypes(include='object').columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in train.select_dtypes(include='object').columns:
    train[col] = le.fit_transform(train[col])
    test[col] = le.fit_transform(test[col])
```

```python
train["Churn"] = train["Churn"].astype(int)
test["Churn"] = test["Churn"].astype(int)
```

```python
X_train = train.drop("Churn", axis=1)
y_train = train["Churn"]

X_test = test.drop("Churn", axis=1)
y_test = test["Churn"]
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,pred))
```
